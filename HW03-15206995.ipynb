```python
import math 
import matplotlib.pyplot as plt 
%matplotlib inline

def displacement(m,t): 
    K = 40
    A = 0.1
    return A * math.sin(math.sqrt(K/m)*t)     
```

```python
weight = float(input("Please input a value for the weight between 1 and 5 kgs:"))
```

```python
t = 0 
ts = []
xs = []

for i in range(51): 
    xs.append(displacement(weight, t))
    ts.append(t)
    
    if((i == 25 or i == 50)): 
        plt.plot(ts, xs, "bo-")
        plt.xlabel("time(s)")
        plt.ylabel("diplacement(s)")
        plt.title("path of %.2f kg mass after %.2f seconds" % (weight, t))
        graph = "displacement_at_%.2f_iterations.png" %i
        plt.savefig(graph)
        
    t += 0.1
```
